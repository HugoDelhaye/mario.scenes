```python
from mario_scenes.load_data import load_scenes_info
import os
import shutil
scenes_info = load_scenes_info()
scenes_info
```

```python
# Select a pattern

pattern = "Stair up"
```

```python
# Find all scenes containing the pattern
scenes_list_from_pattern = []
for scene in scenes_info.iterrows():
    print(scene[1][pattern])
    if scene[1][pattern] == 1.0:
        scenes_list_from_pattern.append(f'w{int(scene[1]["World"])}l{int(scene[1]["Level"])}s{int(scene[1]["Scene"])}.png')
```

```python
# Save them in a new folder and create a zip archive

# Define source and destination directories
source_dir = "../sourcedata/scene_backgrounds"  # Change this to your actual scenes folder if different
dest_dir = f"../outputdata/selected_scenes/{pattern}"

# Create destination directory if it doesn't exist
os.makedirs(dest_dir, exist_ok=True)

# Copy each scene file to the new folder
for scene_file in scenes_list_from_pattern:
    src_path = os.path.join(source_dir, scene_file)
    dst_path = os.path.join(dest_dir, scene_file)
    if os.path.exists(src_path):
        shutil.copy2(src_path, dst_path)
    else:
        print(f"File not found: {src_path}")

# Create a zip archive of the new folder
shutil.make_archive(dest_dir, 'zip', dest_dir)
```
